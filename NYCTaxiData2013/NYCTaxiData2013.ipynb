## Taxi Data

Now that we have all the preliminaries out of the way, let's take a look at our data.  The dataset is too large to pull into a DataFrame as is, so we'll have to work with GQL (Google Query Language) for all the aggregation and summary statistics.

```python
# How many records are we looking at?
gbq.read_gbq("select count(*) as trips from [833682135931:nyctaxi.trip_data]")
```

```python
# What does the data look like
gbq.read_gbq("select * from [833682135931:nyctaxi.trip_data] limit 5")
```

```python
def get_trip_count_by_weekday_df():
    query = """
select
  dayofweek(trips.pickup_datetime) as pickup_dayofweek,
  hour(trips.pickup_datetime) as pickup_hour,
  avg(float(trips.trip_time_in_secs)) as mean_trip_time,
  avg(float(trips.trip_distance)) as mean_trip_distance,
  count(*) as trip_count,
  count(distinct date(trips.pickup_datetime)) as distinct_days
from [833682135931:nyctaxi.trip_data] as trips
where float(trips.trip_distance)<1000 -- remove outliers
and float(trips.trip_time_in_secs)<86400
group by pickup_dayofweek, pickup_hour
order by pickup_dayofweek, pickup_hour"""
    df = gbq.read_gbq(query)
    df['trip_count_per_hour'] = df['trip_count']/df['distinct_days']
    return df

trip_count_by_weekday_df = get_trip_count_by_weekday_df()
trip_count_by_weekday_df.head()
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, ncols=1, figsize=(12,12))

for ax, value_field, title in ((ax0, 'trip_count_per_hour', 'Number of Trips per Hour'),
                        (ax1, 'mean_trip_time', 'Average Trip Time (seconds)'),
                        (ax2, 'mean_trip_distance', 'Average Trip Distance (miles)')):
    df = trip_count_by_weekday_df.pivot(index='pickup_dayofweek', columns='pickup_hour', values=value_field)

    heatmap = ax.imshow(df, origin='upper')
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    #fig.colorbar(heatmap)
    plt.colorbar(heatmap, cax=cax)
    ax.set_yticklabels(df.index)
    ax.set_yticks(np.arange(df.index.size), minor=False)
    ax.set_yticklabels(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'], minor=False)
    ax.set_xticks([0,4,8,12,16,20])
    ax.set_xticklabels(['midnight', '4am', '8am', 'noon', '4pm', '8pm'])
    ax.set_xlim(left=0, right=23)
    ax.set_title('NYC Taxi: {} by Day of Week and Time of Day'.format(title))
```
